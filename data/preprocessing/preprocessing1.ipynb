```python
import pandas as pd
import numpy as np
```

## Line chart

```python
def cleaning(df):
    df.loc[df["RussianUkranianWar"] == "< 1", "RussianUkranianWar"] = "0"
    df.loc[df["IsraeliPalestinianWar"] == "< 1", "IsraeliPalestinianWar"] = "0"
    df.loc[df["DrugWar"] == "< 1", "DrugWar"] = "0"

    return df
```

```python
timeline = pd.read_csv("../csv/timeline_trends.csv")
linechart = cleaning(timeline)
linechart.head()
```

```python
linechart.to_csv("../csv/cleaned/timeline_trends_cleaned.csv", index=False)
```
