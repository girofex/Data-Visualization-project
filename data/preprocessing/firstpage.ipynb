## Stacked bar chart
Data used: [Brookings - Ukraine Index: Tracking developments in the Ukraine war](https://www.brookings.edu/articles/ukraine-index-tracking-developments-in-the-ukraine-war/)

```python
def stacked(df):
    df.rename(columns={'Percentage': 'Russian_Possession'}, inplace=True)
    df['Russian_Possession'] = df['Russian_Possession'].str.replace('%', '')
    df['Russian_Possession'] = df['Russian_Possession'].astype(float)
    df['Ukrainian_Possession'] = 100.0 - df['Russian_Possession']

    return df
```

```python
invasion = pd.read_csv("../csv/russian_possession.csv")
cleaned = stacked(invasion)

invasion.head()
```

```python
invasion.to_csv("../csv/cleaned/stacked.csv", index=False)
```

## Polar chart
Data used: [Eurostat - Imports of natural gas by partner country](https://ec.europa.eu/eurostat/databrowser/view/NRG_TI_GAS__custom_938385/bookmark/table?lang=en&bookmarkId=d84ea630-1f0a-4827-891d-f4a8e930dfe7&c=1620753365581)

```python
def polar(df):
    df_cleaned = df.drop(columns=["STRUCTURE", "STRUCTURE_ID", "STRUCTURE_NAME", "freq", "Time frequency", "siec", "Standard international energy product classification (SIEC)",
                          "partner", "unit", "Unit of measure", "geo", "Geopolitical entity (reporting)",
                          "Time", "Observation value", "OBS_FLAG", "Observation status (Flag) V2 structure", "CONF_STATUS", "Confidentiality status (flag)"])
    df_cleaned = df_cleaned[df_cleaned["Geopolitical entity (partner)"] == "Russia"]

    return df_cleaned
```

```python
csv = pd.read_csv("../csv/gas.csv")
cleaned = polar(csv)

cleaned.head()
```

```python
cleaned.to_csv("../csv/cleaned/polar.csv", index=False)
```
