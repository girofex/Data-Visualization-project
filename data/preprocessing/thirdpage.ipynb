## Libraries

```python
import pandas as pd
import numpy as np
```

## Area chart
Data used: [United Nations - World drug report](https://www.unodc.org/unodc/en/data-and-analysis/world-drug-report-2025-annex.html)

```python
def deaths(df):
    df_cleaned = df.drop(columns=["Code"])
    df_cleaned = df_cleaned.groupby("Year", as_index=False)["DeathRate"].mean()
    
    return df_cleaned
```

```python
csv = pd.read_csv("../csv/drug_deaths.csv")
cleaned = deaths(csv)

cleaned.head()
```

```python
cleaned.to_csv("../csv/cleaned/areachart.csv", index=False)
```

## Pictorial unit chart
Data used: [Our World in Data - Drug use disorder death rate](https://ourworldindata.org/grapher/death-rates-from-drug-use-disorders-who)

```python
def top3(df):
    df = df[df["Year"] == 2021]
    df_cleaned = df.drop(columns=["Code"])
    df_cleaned = df_cleaned.groupby("Entity", as_index=False)["DeathRate"].mean()
    df_cleaned = df_cleaned.sort_values("DeathRate", ascending=False).head(3)
    
    return df_cleaned
```

```python
csv = pd.read_csv("../csv/drug_deaths.csv")
cleaned = top3(csv)

cleaned.head()
```

```python
cleaned.to_csv("../csv/cleaned/pictorial.csv", index=False)
```
