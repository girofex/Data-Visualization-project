## Bar chart

```python
fatalities = pd.read_csv('../csv/fatalities_israelpalestine.csv')

columns_to_keep = ['side_a_new_id', 'side_b_new_id', 'where_description', 'deaths_a', 'deaths_b', 'deaths_civilians', 'deaths_unknown', 'date_start']
fatalities = fatalities[columns_to_keep]
fatalities.sort_values(by='date_start', inplace=True)
fatalities['year'] = pd.to_datetime(fatalities['date_start']).dt.year   #Extract year from date_start

fatalities.head()
```

```python
#New dataframe to store total fatalities per side
fatalities_per_side = pd.concat([pd.Series(fatalities['deaths_a'].sum()), pd.Series(fatalities['deaths_b'].sum()), pd.Series(fatalities['deaths_civilians'].sum())], axis=1)
fatalities_per_side.columns = ['Government of Israel', 'Hamas and armed groups', 'Civilians']
fatalities_per_side.index = ['Total']
fatalities_per_side.to_csv('../csv/cleaned/barchart.csv')
```
