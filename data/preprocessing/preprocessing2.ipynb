## Gas import

```python
def gas(df):
    df_cleaned = df.drop(columns=["STRUCTURE", "STRUCTURE_ID", "STRUCTURE_NAME", "freq", "Time frequency", "siec", "Standard international energy product classification (SIEC)",
                          "partner", "unit", "Unit of measure", "geo", "Geopolitical entity (reporting)",
                          "Time", "Observation value", "OBS_FLAG", "Observation status (Flag) V2 structure", "CONF_STATUS", "Confidentiality status (flag)"])
    #order and take top 10
    return df_cleaned
```

```python
csv = pd.read_csv("../csv/gas.csv")
cleaned = gas(csv)
cleaned.head()
```

```python
cleaned.to_csv("../csv/cleaned/gas_cleaned.csv", index=False)
```
